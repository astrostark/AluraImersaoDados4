Processamento dos dados da coluna valor para poder utilizar o Valor dos imóveis. 

Existem valores de aluguem (por dia, mês e ano), e nesse caso, generalizei como o valor do imóveis, já que eles irão se tornar outliers 

```python
dados["Valor_Int"] = dados["Valor"].str.replace("R","")
dados["Valor_Int"] = dados["Valor_Int"].str.replace("$","")
dados["Valor_Int"] = dados["Valor_Int"].str.replace(" ","")
dados["Valor_Int"] = dados["Valor_Int"].str.replace("/","")
dados["Valor_Int"] = dados["Valor_Int"].str.replace("\n","")
dados["Valor_Int"] = dados["Valor_Int"].str.replace(".","")
dados["Valor_Int"] = dados["Valor_Int"].str.replace("Ano","")
dados["Valor_Int"] = dados["Valor_Int"].str.replace("Mês","")
dados["Valor_Int"] = dados["Valor_Int"].str.replace("Dia","")
dados["Valor_Int"] = dados["Valor_Int"].astype(int)
dados["Valor_Int"]
```

### 1) Criar uma escala de valores em milhões

```python
dados_vendas["Valor_anuncio_milhao"] = dados_vendas["Valor_anuncio_float"].apply(lambda x: x/1000000)
dados_vendas.head()
```

### 2) Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala)

```python
# dados_vendas["Banheiros"].value_counts() # 9=+
# dados_vendas["Quartos"].value_counts() # 8=+
# dados_vendas["Vagas"].value_counts() # 12=+
# dados_vendas["Metragem"].sort_values(ascending = True).tail(50) # 
# dados_vendas["Metragem"].sort_values(ascending = True).head(50) # 10 >= | > 35677
dados_vendas_limpo = dados_vendas[(dados_vendas["Banheiros"] < 9) & (dados_vendas["Quartos"] < 8) & (dados_vendas["Vagas"] < 12) & (dados_vendas["Metragem"] > 10) & (dados_vendas["Metragem"] < 35000)]
dados_vendas_limpo
```

```python
sns.set()
plt.figure(figsize=(10, 8))
sns.set_style("white")
ax = sns.histplot(data = dados_vendas_limpo, x = "Valor_anuncio_milhao", kde = True, color="red")
ax.set_title("Imóveis na cidade de São Paulo")
ax.set_xlabel("Valor do Imóvel em Milhão")
ax.set_ylabel("Quantidade de Imóveis")
plt.xlim((0, 20))
plt.show()
```

### 3) Preço do metro quadrado por bairro e plotar em um gráfico ideal

```python
dados_vendas["valor_metro_quadrado_imovel"] = dados_vendas["Valor_anuncio_float"]/dados_vendas["Metragem"]
dados_vendas
```

```python
dados_vendas_sum = dados_vendas.groupby("Bairro")[["Metragem","Valor_anuncio_float"]].sum()
dados_vendas_sum["valor_metro_quadrado_bairro_soma"] = dados_vendas_sum.apply(lambda x: x["Valor_anuncio_float"]/x["Metragem"], axis=1)
dados_vendas_sum = dados_vendas_sum.reset_index()
dados_vendas_sum
```

```python
dados_vendas_mean = dados_vendas.groupby("Bairro")[["Metragem","Valor_anuncio_float", "valor_metro_quadrado_imovel"]].mean()
dados_vendas_mean["valor_metro_quadrado_bairro_media"] = dados_vendas_mean["valor_metro_quadrado_imovel"]
dados_vendas_mean
```

```python
dados_vendas_m2_bairro = pd.merge(dados_vendas, dados_vendas_sum, left_on="Bairro", right_on="Bairro", suffixes=("", "_to_delete"))
dados_vendas_m2_bairro = pd.merge(dados_vendas_m2_bairro, dados_vendas_mean, left_on="Bairro", right_on="Bairro", suffixes=("", "_to_delete"))
dados_vendas_m2_bairro.drop(columns=["Metragem_to_delete", "Valor_anuncio_float_to_delete", "Valor_anuncio_float_to_delete", "valor_metro_quadrado_imovel_to_delete"],inplace=True)
dados_vendas_m2_bairro
```

```python
dados_vendas_m2_bairro_unique = dados_vendas_m2_bairro[["Bairro","valor_metro_quadrado_bairro_soma", "valor_metro_quadrado_bairro_media"]].drop_duplicates()
dados_vendas_m2_bairro_unique
```

```python
dados_vendas_m2_bairro_sort = dados_vendas_m2_bairro_unique.sort_values(by="valor_metro_quadrado_bairro_soma", ascending=False).head(10)

plt.figure(figsize=(25,5))

ax = sns.barplot(x=dados_vendas_m2_bairro_sort.Bairro, y=dados_vendas_m2_bairro_sort.valor_metro_quadrado_bairro_soma)

plt.title("Valor do metro quadrado em SP - Top 10 - Cálculo por Soma", fontsize=30, fontweight="bold")
plt.ylabel("Valor", fontsize=20, fontweight="bold")
plt.xlabel("Bairro", fontsize=20, fontweight="bold")
plt.tick_params(axis='y', labelsize=20)
plt.tick_params(axis='x', labelsize=15)

plt.show()
```

```python
dados_vendas_m2_bairro_sort = dados_vendas_m2_bairro_unique.sort_values(by="valor_metro_quadrado_bairro_media", ascending=False).head(10)

plt.figure(figsize=(25,5))

ax = sns.barplot(x=dados_vendas_m2_bairro_sort.Bairro, y=dados_vendas_m2_bairro_sort.valor_metro_quadrado_bairro_media)

plt.title("Valor do metro quadrado em SP - Top 10 - Cálculo por Média", fontsize=30, fontweight="bold")
plt.ylabel("Valor", fontsize=20, fontweight="bold")
plt.xlabel("Bairro", fontsize=20, fontweight="bold")
plt.tick_params(axis='y', labelsize=20)
plt.tick_params(axis='x', labelsize=15)

plt.show()
```
